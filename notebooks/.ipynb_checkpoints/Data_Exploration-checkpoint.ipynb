```python
import pandas as pd
file_path = r"C:\Users\keert\OneDrive\Desktop\Shopping_Deals_Recommeneder\dataset\E-commerce Customer Behavior Dataset.xlsx"
df = pd.read_excel(file_path)

# Displaying first 5 rows
df.head()
```

```python
#Check for Missing Values
df.isnull().sum()
```

```python
# Fill Missing Values with Most Frequent Category (Mode)
df['Satisfaction Level'] = df['Satisfaction Level'].fillna(df['Satisfaction Level'].mode()[0])
```

```python
df.isnull().sum()
```

```python
#Check for Duplicate Rows
df.duplicated().sum()
```

```python
# Check Data Types
df.info()
```

```python
#Basic Statistics
df.describe()
```

```python
#AGE DISTRIBUTION

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.histplot(df["Age"], bins=10, kde=True, color="skyblue")
plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#SPENDING PATTERNS

plt.figure(figsize=(8, 5))
sns.histplot(df["Total Spend"], bins=10, kde=True, color="lightcoral")
plt.title("Total Spend Distribution")
plt.xlabel("Total Spend ()")
plt.ylabel("Count")
plt.show()
```

```python
#Spending by Membership Type

plt.figure(figsize=(8, 5))
sns.boxplot(x="Membership Type", y="Total Spend", data=df, palette="coolwarm")
plt.title("Spending Behavior by Membership Type")
plt.xlabel("Membership Type")
plt.ylabel("Total Spend ()")
plt.show()
```

```python
#Satisfaction Levels

plt.figure(figsize=(6, 4))
sns.countplot(x="Satisfaction Level", hue="Satisfaction Level", data=df, palette="viridis", dodge=False)
plt.title("Distribution of Satisfaction Levels")
plt.xlabel("Satisfaction Level")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend([], [], frameon=False)  # Remove duplicate legend
plt.show()

```

```python
#Satisfaction Level vs. Membership Type

plt.figure(figsize=(6, 4))
sns.countplot(x="Membership Type", hue="Satisfaction Level", data=df, palette="coolwarm")
plt.title("Satisfaction Level Across Membership Types")
plt.xlabel("Membership Type")
plt.ylabel("Count")
plt.legend(title="Satisfaction Level")
plt.show()
```

```python
#Satisfaction vs. Total Spend

plt.figure(figsize=(6, 4))
sns.boxplot(x="Satisfaction Level", y="Total Spend", data=df, palette="Set2")
plt.title("Total Spend vs. Satisfaction Level")
plt.xlabel("Satisfaction Level")
plt.ylabel("Total Spend ()")
plt.show()
```
